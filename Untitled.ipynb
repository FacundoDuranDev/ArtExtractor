```python
import requests
from bs4 import BeautifulSoup
import os
import time

# Rango de IDs de prueba
start_id = 1870
end_id = 1880
delay = 1  # Pausa entre solicitudes para evitar sobrecargar el servidor

# Crear un directorio para guardar las imágenes
os.makedirs('imagenes_obras', exist_ok=True)

# Base URL de las obras
base_url = "https://www.bellasartes.gob.ar/coleccion/obra/"

for obra_id in range(start_id, end_id + 1):
    url = f"{base_url}{obra_id}/"
    response = requests.get(url)
    
    if response.status_code == 200:
        print(f"Obra encontrada en {url}")
        
        # Analizar el contenido HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extraer el título de la obra
        titulo_tag = soup.find('h1')
        titulo = titulo_tag.text.strip() if titulo_tag else "Título no encontrado"
        
        # Extraer la URL de la imagen
        img_tag = soup.find('img')
        img_url = img_tag['src'] if img_tag and 'src' in img_tag.attrs else None
        
        # Mostrar título y URL de la imagen
        print(f"Título: {titulo}")
        print(f"URL de la imagen: {img_url}")
        
        if img_url:
            # Asegurarse de que la URL sea completa
            if img_url.startswith('/'):
                img_url = f"https://www.bellasartes.gob.ar{img_url}"
            
            # Descargar la imagen
            img_response = requests.get(img_url)
            if img_response.status_code == 200:
                # Guardar la imagen
                img_filename = os.path.join('imagenes_obras', f"{obra_id}.jpg")
                with open(img_filename, 'wb') as f:
                    f.write(img_response.content)
                print(f"Imagen guardada como {img_filename}")
            else:
                print(f"Error al descargar la imagen: {img_response.status_code}")
    else:
        print(f"Obra no encontrada para ID {obra_id}")
    
    # Pausa entre solicitudes
    time.sleep(delay)
```

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, HTML
# URL de la página que deseas analizar
url = "https://www.bellasartes.gob.ar/coleccion/obra/1876/"  # Cambia esto a la URL que desees

# Hacer la solicitud HTTP
response = requests.get(url)
base_url_img = "https://www.bellasartes.gob.ar/"

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    titulo_tag = soup.find('h1', class_='h3 text-white')
    titulo = titulo_tag.text.strip() if titulo_tag else "Título no encontrado"
    print(f"Título: {titulo}")
    
    details_div = soup.find('dl', class_='row mt-3')
    
    if details_div:
        li_elements = details_div.find_all('li')
        
        for li in li_elements:
            print(li.text.strip())  # Mostrar el texto del elemento <li>
    else:
        print("No se encontró la lista de detalles.")

    img_tag = soup.find('a', {'data-fancybox': 'gallery'})
    if img_tag and 'href' in img_tag.attrs:
        img_url = img_tag['href']

        # Descargar la imagen
        html_img = f'<img src="{base_url_img+img_url}" width="400" height="300" />'  # Cambia 400 y 300 por los valores que desees

        # Mostrar la imagen usando HTML
        display(HTML(html_img))
        
else:
    print(f"Error al acceder a la URL: {response.status_code}")
```

```python
import base64

# IDs de obras que quieres convertir a Base64
ids_obras = [1872, 1873, 1874, 1876, 1879, 1880]

# Función para convertir un número en Base64
def convertir_id_a_base64(id_num):
    # Convertir el número a bytes (UTF-8) para codificación
    id_bytes = str(id_num).encode("utf-8")
    # Codificar en Base64
    id_base64 = base64.b64encode(id_bytes).decode("utf-8")
    return id_base64

# Convertir y mostrar cada ID en Base64
for id_obra in ids_obras:
    id_base64 = convertir_id_a_base64(id_obra)
    print(f"ID {id_obra} en Base64: {id_base64}")
```
